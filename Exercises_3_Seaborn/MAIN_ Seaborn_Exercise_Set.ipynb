<ul>
<li>Create a new column: <b>"liters_per_100km"</b> , which converts the mpg to liters per 100 km</li>
<ul>
<li>You can Google the conversion formula easily: "miles per gallon to liters per 100km" or "How do you convert MPG to l 100km?"
</li>
<li>remove the original mpg column after this</li>
</ul>

</ul>

```python
# Loading the dataset
df = sns.load_dataset('mpg')
```

```python
# Function to convert gallons to litre / 100km
def convert_mpg_100km(row):
    conversion = 235.215 / (row)
    return conversion

# Applying the function to a new column called litre_100km
df['litre_100km'] = df['mpg'].apply(convert_mpg_100km)

# Rounding to 2 decimals for clarity.
df['litre_100km'] = df['litre_100km'].round(2)

# Dropping the mpg column
df = df.drop('mpg', axis=1)
df.head(3)
```

<li>Remove the "name" -column</li>

```python
# Dropping the 'name' column.
df = df.drop('name', axis=1)
df.head(3)
```

<li>If you take a look at the data, you will notice three columns have NaN 
values. You can either remove these columns all together, or you can fill
the missing values with average values of that column. For example, for

```python
# Loading the dataframe.
df = pd.read_csv('groceries.csv')
df

# We notice we have 360 rows, that's not a lot of data.
```

```python
# Handle incomplete rows here, justify your approach on how you handled them.

# Checking the amount of null rows.
df.isna().sum()
```

```python
# Using imputation to replace the NaN values with the average of the column.

# 50% of the sunflower-oil column is missing. That's a lot of missing rows. let's drop the column.
df = df.drop('Sunflower-oil', axis = 1)

# Around 10-20% is missing in these columns, let's imputate the average to fix them.
df['Fish'].fillna((df['Fish'].mean()), inplace=True)
df['Pork'].fillna((df['Pork'].mean()), inplace=True)

# We only have 360 rows, so 39 rows is still quite a lot, so we imputated instead of dropping the rows.
```

```python
# Re-checking NaN values.
df.isna().sum()
```

<ul>
<li>In this data, the date column is a bit difficult to use, since it's not
completely in numerical format. Split the Month-column so, that you
have two different columns: Month and Year</li>
<ul>
<li>For month, use a numeric format 1-12</li>
<li>For year, use the full year 1990-2020</li>
<li>Check out the examples in Moodle, this one is a bit tricky, but

```python
# A different approach
# Converting the 'Month' column into datetime, using the format parameter.
df['Month'] = pd.to_datetime(df["Month"], format="%b-%y")

# Dividing the datetime to Year and Month
df['Year'] = df['Month'].dt.year
df['Month'] = df['Month'].dt.month
```

```python
# Moving the Year to the first index position. It makes more sense that way.
column_to_move = df.pop('Year')
df.insert(0, 'Year', column_to_move)

df.head(2)
```
